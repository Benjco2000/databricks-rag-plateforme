```python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

# Liste toutes les apps et leur état
apps = w.apps.list()
for app in apps:
    print(f"App : {app.name}")
    print(f"  Status     : {app.status}")
    print(f"  URL        : {app.url}")
    print(f"  Active dep : {app.active_deployment}")
    print()
```
